```python
# Step 1: Import Libraries
!pip install pandas seaborn matplotlib --quiet  
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Step 2: Load Dataset 
iris = sns.load_dataset('iris')  # Direct from Seaborn!
# step 3: Basic Inspection
print("Shape of Dataset:", iris.shape)
print("\nFirst 5 Rows:\n", iris.head())
print("\n Column Names:", iris.columns.tolist())

# Step 4: Summary Statistics
print("\nðŸ’¡ Dataset Info:")
iris.info()
print("\n Descriptive Stats:\n", iris.describe())
```

```python
# Step 5: VISUALIZATION
# Scatter Plot
sns.pairplot(iris, hue='species', palette='viridis')
plt.suptitle("Feature Relationships (Scatter Plots)", y=1.02)
plt.show()
```

```python
# step 6: Histograms
iris.hist(figsize=(12, 8), color='purple', edgecolor='black')
plt.suptitle("Value Distributions (Histograms)", y=1.02)
plt.show()
```

```python
#step 7: Box Plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris.drop('species', axis=1), palette='magma')
plt.title("Box Plots (Spot the Outliers!)")
plt.show()
```
